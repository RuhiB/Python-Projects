```python
# Importing the required libraries

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading the CSV file into a DataFrame

covid_df=pd.read_csv("python_notebook/country_vaccinations.csv")
```

```python
covid_df.head()
```

```python
covid_df.info()
```

```python
# Handeling all the missing data

covid_df.fillna(0,inplace=True)
```

```python
covid_df.head()
```

```python
#1)Which country is using what vaccine?
```

```python
data_df=pd.DataFrame(covid_df.groupby('country')['vaccines'].unique())
```

```python
data_df
```

```python
#2) In which country the vaccination program is more advanced?

TVacc = covid_df['total_vaccinations'].groupby(covid_df['country']).max().sort_values(ascending=False)
```

```python
TVacc_df = pd.DataFrame(TVacc)
```

```python
# Lets take a look at first 20 countries

TVacc_df[:20]
```

```python
TVacc_df[:20].plot(kind='bar',figsize=(12,6))
```

```python
#3) Where are more people vaccinated per day, in terms of percent from entire population?

TVPH_df = pd.DataFrame(covid_df.groupby('country')['total_vaccinations_per_hundred'].max().sort_values(ascending=False)[:20])
```

```python
TVPH_df.plot(kind='bar',figsize=(12,6))
```

```python
# Considering number of people vaccinated per 100 people, Gibraltar is the leading country in vaccination.
# Gibraltar has more vaccinated population per 100 people.
```

```python
#4) Where are more people daily vaccinated per million?
```

```python
TVPM = covid_df['daily_vaccinations_per_million'].groupby(covid_df['country']).max().sort_values(ascending=False)[:20]
```

```python
TVPM_df=pd.DataFrame(TVPM).plot(kind='bar',figsize=(12,6))
```

```python
# Bhutan has more daily vaccinations per million.
```

```python
#5) Which country has highest percentage of people fully immunised to total population?

PFVH = pd.DataFrame(covid_df.groupby('country')['people_fully_vaccinated_per_hundred'].max().sort_values(ascending=False)[:20])
```

```python
PFVH.plot(kind='bar',figsize=(12,6))
```

```python
# Gibraltar is again the leading nation where most of the population has received both the doses of vaccinations.
# If a person receives a single dose of vaccine, where two doses are necessary, that is not considered as fully vaccinated.
```

```python
#6) In which month where the number of vaccinations highest?
```
